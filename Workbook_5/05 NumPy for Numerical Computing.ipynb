[あ]

- `class`
  - a code template which contains a set of related of functions, constants, and variables
- `property` or `attribute`
  - a variable inside a class
- `method`
  - a function inside a class
- `object`
  - a specific instance of a class
- `module`  
  - a collection of pre-written code  that gives a set of related functionalities
- `package`
  - a collection of related modules. Also called a `library`

### Why use arrays over lists?

So far the `numpy.ndarray` looks awefully much like a Python list (or nested list). Why not simply use Python lists for computations instead of creating a new array type?

There are several reasons:

* Python lists are very general. They can contain any kind of object. They are **dynamically typed**. They do not support mathematical functions such as matrix and dot multiplications, etc. Implementing such functions for Python lists would not be very efficient because of the dynamic typing.
* Numpy arrays are **statically typed** and **homogeneous**. The type of the elements is determined when the array is created.
* Numpy arrays are memory efficient.
* Because of the static typing, fast implementation of mathematical functions such as multiplication and addition of `numpy` arrays can be implemented in a compiled language (C and Fortran is used).


**In summary**: numpy arrays are used over lists for computing efficiency and power. This is possible because of the certainty that the array will always contain numeric data. [あ]

## 3.2 Index slicing

Index slicing is the technical name for the syntax `M[lower:upper:step]` to extract part of an array:

Similar to slicing of `Strings`, the element at `upper` is __not included__ in the slice.
[あ]

````python
v = np.array([1,2,3,4,5])
v
````

````python
v[1:3]
````

We can omit any of the three parameters in `M[lower:upper:step]`:

````python
v[::] # lower, upper, step all take the default values (step=1)
````

````python
v[::2] # step is 2, lower and upper defaults to the beginning and end of the array
````

````python
v[:3] # first three elements
````

````python
v[3:] # elements from index 3
````

Negative indices counts from the end of the array (positive index from the begining):

````python
v = np.array([1,2,3,4,5])
````

````python
v[-1] # the last element in the array
````

````python
v[-3:] # the last three elements
````

Index slicing works exactly the same way for multidimensional arrays:

````python
M = np.array([[ 1,  2,  3,  4],
              [11, 12, 13, 14],
              [21, 22, 23, 24],
              [31, 32, 33, 34]])
M
````

````python
# strides
M[::2, ::2]
````

## 3.4 Copy and "deep copy"

To achieve high performance, `assignment statements` in Python usually do not copy the underlying objects. This is important for example when objects are passed between functions, to avoid an excessive amount of memory copying when it is not necessary (technical term: `pass by reference`).

You are just passing a link (the location of the data), not the actual object and its contents

````python
A = np.array([[1, 2], [3, 4]])
A
````

````python
B = A
B
# now B is referring to the same array data as A
````

````python
# changing B affects A
B[0,0] = 10
B
````

````python
A
````

If we want to avoid this behavior, so that when we get a new completely independent object `B` copied from `A`, then we need to do a so-called __deep copy__ using the function `copy`:

````python
B = np.copy(A)
B
````

````python
# now, if we modify B, A is not affected
B[0,0] = -5
B
````

````python
A
````

#あ: Python has created a new object `B`, copying the contents of `A` to it.

## Using `%timeit` to inspect runtime

* an Ipython magic command to measure the execution time taken by the given line of code
* Magic commands are enhancements on top of normal Python syntax
  + prefix by `%` character

````python
# displaying all available magic commands
%lsmagic
````

````python
%timeit
import numpy as np
````

#### Timing with `%timeit`

````python
%timeit np.random.rand(1000)
````

The output above shows the average of timing statistics such as the `mean`, `standard deviation` of time. These indicate the summary of the runtime considering each of the multiple runs.

+ `mean` = `9.75 µs`
+ `standard deviation` = `± 57 ns`

Multiple runs and loops were also generated. This provides a more accurate representation of the actual runtime rather than relying on just one iteration to calculate the runtime.

#### Specifying the number of runs/ loops

+ Setting the number of runs (`-r`) and/or loops (`-n`).

````python
# Set the runs to 2 (-r2)
# Set the loops to 10 (-n10)
%timeit -r2 -n10 np.random.rand(1000)
````

The statement above was executed **20 times** in order to estimate runtime (2 runs each with 10 loops)

### Running multiple lines of codes with `%%timeit`

````python
%%timeit
num = []
for x in range(10):
  num.append(x)
````

### Saving `%timeit` output

+ saving the output into the variable `-o`

````python
run_time = %timeit -o rand_nums = np.random.rand(1000)
````

Allows to view the time for each run.

````python
run_time.timings
````

Allows to view the best time for all runs

````python
run_time.best
````

Allows to view the worst time for all runs

````python
run_time.worst
````
